## Q-2. Write an SQL query to fetch duplicate records having matching data in some fields of a table.

```python
#finding duplicate last names
cursor.execute('select last_name, count(last_name) from employees group by last_name having count(last_name) > 1')
cursor.fetchall()
```

```python
#finding duplicate first name
cursor.execute('select first_name, count(first_name) from employees group by first_name having count(first_name) > 1')
cursor.fetchall()
```

```python
cursor.execute('select first_name, last_name, gender, count(*) from employees group by first_name, last_name, gender having count(*)>1')
cursor.fetchall()
```
